# Dimentionality Reduction

```python
# Reduce variable in training data
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=['age', 'gender', 'chest_pain', 'high_blood_pressure',
       'irregular_heartbeat', 'shortness_of_breath', 'fatigue_weakness',
       'dizziness', 'swelling_edema', 'neck_jaw_pain', 'excessive_sweating',
       'persistent_cough', 'nausea_vomiting', 'chest_discomfort',
       'cold_hands_feet', 'snoring_sleep_apnea', 'anxiety_doom',
       'stroke_risk_percentage', 'at_risk']
a=sc.fit_transform(df[x])
pca=PCA(n_components=2)
pca_r=pca.fit_transform(a)
df['pca']=pca_r[:,0]
df["pca2"]=pca_r[:,1]
```

 Principal Component Analysis (PCA) after standardizing the data using StandardScaler. The main goal is to reduce the dataset's dimensionality while preserving as much variance as possible.

# Scaling


```python
# Numerical features in a dataset to a common scale
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
df[['age', 'gender', 'chest_pain', 'high_blood_pressure',
       'irregular_heartbeat', 'shortness_of_breath', 'fatigue_weakness',
       'dizziness', 'swelling_edema', 'neck_jaw_pain', 'excessive_sweating',
       'persistent_cough', 'nausea_vomiting', 'chest_discomfort',
       'cold_hands_feet', 'snoring_sleep_apnea', 'anxiety_doom',
       'stroke_risk_percentage', 'at_risk','pca','pca2']]=mm.fit_transform(df[['age', 'gender', 'chest_pain', 'high_blood_pressure',
       'irregular_heartbeat', 'shortness_of_breath', 'fatigue_weakness',
       'dizziness', 'swelling_edema', 'neck_jaw_pain', 'excessive_sweating',
       'persistent_cough', 'nausea_vomiting', 'chest_discomfort',
       'cold_hands_feet', 'snoring_sleep_apnea', 'anxiety_doom',
       'stroke_risk_percentage', 'at_risk','pca','pca2']])
```

This code normalizes the dataset using Min-Max Scaling. The goal is to scale all feature values between 0 and 1, making them comparable and suitable for machine learning models.process of transforming data values to a specific range

```python
# model evaluation
from sklearn.model_selection import train_test_split
y=df["stroke_risk_percentage"]
x=df.drop(columns=['stroke_risk_percentage'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
print(f'x_train:{x_train.shape}\ny_train:{y_train.shape}\nx_test:{x_test.shape}\ny_test:{y_test.shape}')
```

This code splits the dataset into training and testing sets to evaluate the performance of machine learning models.

```python
print(x_train)
y_train
# print(x_test)
y_test
```

# Initial Modelling and hypothesis

```python
from sklearn.svm import SVR
svm_model = SVR()
# fit the model to the train data
svm_model.fit(x_train,y_train)
# predict the target value of the test data
svm_prediction = svm_model.predict(x_test)
print("SVM_Prediction",svm_prediction)# y_test predict the value

# calculating error
from sklearn.metrics import mean_absolute_error
svm_mae=mean_absolute_error(y_test,svm_prediction)
print("Mean absolute error of SVM = ",svm_mae)

from sklearn.metrics import mean_squared_error
import math # Import the math module

# rmse of the data
svm_mse=mean_squared_error(y_test,svm_prediction)
svm_rmse=math.sqrt(svm_mse)  # Use math.sqrt to calculate the square root
print("Root Mean Squared Error of SVM",svm_rmse)

# to find the accuracy of the above dataset
from sklearn.metrics import r2_score
r=r2_score(y_test,svm_prediction)
r
```

Support Vector Regression (SVR) model, makes predictions, and evaluates the model's performance using different error metrics.SVR is used for predicting continuous values.It works by finding a function that fits the data while minimizing prediction errors.
